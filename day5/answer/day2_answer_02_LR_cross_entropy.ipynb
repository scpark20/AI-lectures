```python
import tensorflow as tf

## data 
x_data = [[5.], [30.], [95.], [100.], [265.], [270.], [290.], [300.],[365.]]
y_data = [[0.], [0.],  [0.], [0.], [1.], [1.], [1.], [1.], [1.]]
test_data= [[7.]]
test_data2= [[80.]]
test_data3= [[110.]]
test_data4= [[180.]]
test_data5= [[320.]]

## tf.keras  perceptron  .
model = tf.keras.Sequential() ##  
model.add(tf.keras.layers.Dense(1, input_dim=1,activation='sigmoid')) #   add .
model.summary()

#  loss,   
optimizer=tf.keras.optimizers.SGD(learning_rate=0.0007) ###   global min     .
loss=tf.keras.losses.binary_crossentropy
metrics=tf.keras.metrics.binary_accuracy

#   -   loss        
model.compile(loss=loss, optimizer=optimizer, metrics=[metrics])

#  
model.fit(x_data, y_data, epochs=5000, batch_size=9)

#  .
print(" test data [7.]   : ", model.predict(test_data))
print(" test data [80.]   : ", model.predict(test_data2))
print(" test data [110.]   : ", model.predict(test_data3))
print(" test data [180.]   : ", model.predict(test_data4))
print(" test data [320.]   : ", model.predict(test_data5))
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = list(range(400))
y = []
for i in x:
    y_pred = model.predict([[i]])
    y.append(y_pred[0][0])

plt.plot(x, y)
plt.ylabel('y_pred')
plt.ylabel('X')
plt.show()
```
