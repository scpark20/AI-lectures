## Single Perceptron

```python
import tensorflow as tf

## data 
x_data = [[0.,0.], [0.,1.], [1.,0.],[1.,1.]]
y_data = [[0.], [1.], [1.], [0.]]
test_data=[[0.5, 0.5]]

## tf.keras  perceptron  .
model = tf.keras.Sequential() ##  
model.add(tf.keras.layers.Dense(1, input_dim=2,activation='sigmoid')) #   add .
model.summary()

#  loss,   
optimizer=tf.keras.optimizers.SGD(learning_rate=0.005) ###   global min     .
loss=tf.keras.losses.binary_crossentropy  ##     .
metrics=tf.keras.metrics.binary_accuracy ###    

#  
model.compile(loss=loss, optimizer=optimizer, metrics=[metrics])

#  
model.fit(x_data, y_data, epochs=3000, batch_size=4)
```

## Multi-layer Perceptron (Sequantial API)

```python
import tensorflow as tf

## data 
x_data = [[0.,0.], [0.,1.], [1.,0.],[1.,1.]]
y_data = [[0.], [1.], [1.], [0.]]
test_data=[[0.5, 0.5]]


## tf.keras  perceptron  .
model = tf.keras.Sequential() ##  
model.add(tf.keras.layers.Dense(4, input_dim=2, activation='sigmoid')) #   add .
model.add(tf.keras.layers.Dense(5, activation='sigmoid'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid')) #   add .
model.summary()

#  loss,   
optimizer=tf.keras.optimizers.SGD(learning_rate=0.5) ###   global min     .
loss=tf.keras.losses.binary_crossentropy  ##     .
metrics=tf.keras.metrics.binary_accuracy ###    

#  
model.compile(loss=loss, optimizer=optimizer, metrics=[metrics])

#  
model.fit(x_data, y_data, epochs=2000, batch_size=4)
```

## Multi-layer Perceptron (Functional API)

```python
import tensorflow as tf
import numpy as np

## data 
x_data = [[0.,0.], [0.,1.], [1.,0.],[1.,1.]]
y_data = [[0.], [1.], [1.], [0.]]
test_data=[[0.5, 0.5]]


## tf.keras  perceptron  .
input_Layer = tf.keras.layers.Input(shape=(2,))   ##
#    ( None, 2)  None default         .
## (2,) -> 1  [ x1, x2 ]
##   data  [[0.,0.], [0.,1.], [1.,0.],[1.,1.]] ->  1  (2,)  4    
### (batch_size( ) ,     (  )) -> ( None,    )
##     ?
###  (w, h, c)
## ( None, w, ,h, c )  tf.keras.layers.Input(shape=(w,h,c))

x1 = tf.keras.layers.Dense(4, activation='sigmoid')(input_Layer)
x2 = tf.keras.layers.Dense(5, activation='sigmoid')(x1)
Out_Layer= tf.keras.layers.Dense(1, activation='sigmoid')(x2)
model = tf.keras.Model(inputs=[input_Layer], outputs=[Out_Layer])
model.summary()

#  loss,   
optimizer=tf.keras.optimizers.SGD(learning_rate=0.5) ###   global min     .
loss=tf.keras.losses.binary_crossentropy  ##     .
metrics=tf.keras.metrics.binary_accuracy ###    

#  
model.compile(loss=loss, optimizer=optimizer, metrics=[metrics])

#  
model.fit(x_data, y_data, epochs=2000, batch_size=4)
# model.fit(x_data,y_data, epochs=1000,)
```
