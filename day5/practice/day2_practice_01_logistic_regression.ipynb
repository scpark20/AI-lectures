```python
import tensorflow as tf

## data 
x_data = [[2.], [4.], [6.], [8.], [10.], [12.], [14.]]
y_data = [[0.], [0.], [0.], [1.], [1.], [1.], [1.]]
test_data= [[3.]]
test_data2= [[7.]]
test_data3= [[17.]]
test_data4= [[60.]]

## tf.keras  perceptron  .
model = tf.keras.Sequential() ##  
model.add(tf.keras.layers.Dense(1, input_dim=1, activation='sigmoid')) #   add .
model.summary()

#  loss,   
optimizer=tf.keras.optimizers.SGD(learning_rate=0.01)   ###   global min     .
loss=tf.keras.losses.mean_squared_error   ##     . mse mean squre error ( - )^2  
metrics=tf.keras.metrics.binary_accuracy  ###    

#   -   loss        
model.compile(loss=loss, optimizer=optimizer, metrics=[metrics])

#  
model.fit(x_data, y_data, epochs=1500, batch_size=2)

#  .
print(" test data [3.]   : ", model.predict(test_data))
print(" test data [7.]   : ", model.predict(test_data2))
print(" test data [17.]   : ", model.predict(test_data3))
print(" test data [60.]   : ", model.predict(test_data4))

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = list(range(50))
y = []
for i in x:
    y_pred = model.predict([[i]])
    y.append(y_pred[0][0])

plt.plot(x, y)
plt.ylabel('y_pred')
plt.ylabel('X')
plt.show()
```
