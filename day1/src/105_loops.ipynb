# 1. while문

```python
while 조건 :
    수행할 문장1
    수행할 문장2
```

```python
# 10이하의 짝수 프린트하기
i = 1

while i <= 10 :
    if i % 2 == 0 :
        print(i)
    i += 1
```

## break

```python
# 100번째 방문자 찾기
i = 90

while i :
    i += 1
    if i == 100 :
        print("축하합니다. %d번째 방문자입니다." % i)
        break
print("감사합니다. 이벤트가 종료되었습니다.")
```

## continue

```python
i = 0

while i < 11 :
    i += 1
    if i == 6 :
        continue
    if i % 2 == 0 : 
        print(i)
```

# 2. for문

```python
for 변수 in range(변수가 속한 자료형, 혹은 변수의 범위) :
    수행해야할 문장1
    수행해야할 문장2
```

## 변수에 범위를 지정하기
### 1. range

```python
for x in range(0,5) :
    print(x)
```
