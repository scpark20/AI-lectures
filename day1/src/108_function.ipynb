## return 사용하기

```python
def myfunction3(a) :
    a += 1
    print(a)
    return a
```

```python
b = myfunction3(a)
```

```python
b
```

```python
a = 1
```

## global 사용하기

```python
def myfunction4():
    global a
    a += 1
    print(a)
```

```python
myfunction4()
```

```python
a
```

잠깐! global을 사용하기 전에 반드시 고려해주세요!

```python
for _ in range(5) :
    c = myfunction3(a)
    print('함수의 결과 : %d' %c )
    print('a의 값 : %d' %a)
```

```python
for _ in range(5) :
    myfunction4()
    print('a의 값 : %d' %a)
```
