# if / else

```python
if 조건 :
    수행 할 문장1
    수행 할 문장2
else :
    수행 할 문장3
    수행 할 문장4
```

```python
money = True

if money :
    print("여행을 간다")
else :
    print("집에서 쉰다")
```

# if / elif / else

```python
if 조건1 :
    수행 할 문장1
elif 조건2 :
    수행 할 문장2
else :
    수행 할 문장3
```

```python
money = 20000

if money < 5000 :
    print('라면을 먹는다')
elif 5000 < money < 25000 :
    print('치킨을 먹는다')
elif 25000 < money < 50000 :
    print('삼겹살을 먹는다')
else :
    print('소고기를 먹는다')
```
