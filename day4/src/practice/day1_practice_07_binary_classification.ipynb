```python
import tensorflow as tf

## data 
x_data = [[0.,0.], [0.,1.], [1.,0.],[1.,1.]]
y_data = [[0.], [0.], [0.], [1.]]
test_data=[[0.3, 0.3]]

## tf.keras  perceptron  .
model = tf.keras.Sequential() ##    sequential  .      .
model.add(tf.keras.layers.Dense(1, input_dim=2)) #   add  . ,     2, perceptron 1.
model.summary() ##   

#  loss,   
optimizer=tf.keras.optimizers.SGD(lr=0.001)  ###   global min     .
loss=tf.keras.losses.mean_squared_error  ##     . mse mean squre error ( - )^2  
metrics=tf.keras.metrics.binary_accuracy ###    

#  
model.compile(loss=loss, optimizer=optimizer, metrics=[metrics])

#  
model.fit(x_data, y_data, epochs=2000, batch_size=4)

#  .
print(model.weights)
print(" test data [0.3, 0.3]   : ", model.predict(test_data))
if model.predict(test_data)>0.5:
    print("  " )
else:
    print("  ")
```

```python
test_data=[[0.8, 0.8]]
print(" test data [0.8, 0.8]   : ", model.predict(test_data))
if model.predict(test_data)>0.5:
    print("  " )
else:
    print("  ")
```
