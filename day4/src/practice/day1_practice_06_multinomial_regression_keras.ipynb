```python
import tensorflow as tf
## data 
x_data = [[2.,0.], [4.,4.], [6.,2.],[8.,3.]]
y_data = [[81], [93], [91], [97]]
test_data=[[5.,5.]]
print(len(x_data),len(x_data[1]))  #  , 

## tf.keras  perceptron  .
model = tf.keras.Sequential() ##    sequential  .      .
model.add(tf.keras.layers.Dense(1, input_dim=2)) #   add  . ,     2, perceptron 1.
model.summary() ##   

#  loss,   
optimizer=tf.keras.optimizers.SGD(lr=0.01)  ###   global min     .
loss=tf.keras.losses.mse  ##     . mse mean squre error ( - )^2  
metrics=tf.keras.metrics.mae ###     mse mean_absolute_error | - |  

#  
model.compile(loss=loss, optimizer=optimizer, metrics=[metrics])

#  
model.fit(x_data, y_data, epochs=2000, batch_size=4)

#  .
print(model.weights)
print(" test data [5., 5.]   : ", model.predict(test_data))
```
