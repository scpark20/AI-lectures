```python
import tensorflow as tf

## data 
x_data =[[1.],[2.],[3.],[4.]]
y_data =[[1.],[3.],[5.],[7.]]

##  0,  1     .
#        w,b .
W=tf.Variable(tf.random.normal((1,1),mean=0, stddev=1.0))
b=tf.Variable(tf.random.normal((1,1),mean=0, stddev=1.0))
lr=tf.constant(0.01)

for i in range(2000):  ## 
    total_error = 0

    for j in range(len(x_data)): ##  1
        ## data * weight
        WX =tf.matmul([x_data[j]], W)

        ## bias add
        y_hat = tf.add(WX, b)

        ##  Y  error 
        error = tf.subtract(y_data[j], y_hat) ## (true - prediction)


        ##  W b .
        ##  
        diff_W = tf.matmul(error,[x_data[j]])#error*x 
        diff_b = error

        ##     
        diff_W = tf.multiply(lr, diff_W)
        diff_b = tf.multiply(lr, diff_b)  #  lr * (error)

        ## w, b 
        W=tf.add(W, diff_W) # w + lr * x * (error)
        b=tf.add(b, diff_b) # b + lr * (error)
        #######

        ##  .
        visual_error=tf.square(error)
        total_error = total_error + visual_error

    ##    error 
    print("epoch: ", i, "error : ", total_error/len(x_data))

##   W b  
print("W : ", W)
print("b : ", b)
print("input 3 : ", tf.add(tf.multiply(3, W), b))
print("input 5 : ", tf.add(tf.multiply(5, W), b))
```
