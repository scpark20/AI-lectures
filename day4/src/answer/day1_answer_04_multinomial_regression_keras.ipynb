```python
import tensorflow as tf

## data 
x_data = [[2.,0.,7.], [6.,4.,2.], [5.,2.,4.],[8.,4.,1]]
y_data = [[75], [95], [91], [97]]
test_data=[[5.,5.,5.]]
print(len(x_data),len(x_data[1]))  #  , 

## tf.keras  perceptron  .
model = tf.keras.Sequential() ##    sequential  .      .
model.add(tf.keras.layers.Dense(1, input_dim=3)) #   add  . ,     3, perceptron 1.
model.summary() ##   

#  loss,   
optimizer=tf.keras.optimizers.SGD(lr=0.002)  ###   global min     .
loss=tf.keras.losses.mse  ##     . mse mean squre error ( - )^2  
metrics=tf.keras.metrics.mae ###     mse mean_absolute_error | - |  

#  
model.compile(loss=loss, optimizer=optimizer, metrics=[metrics])

#  
model.fit(x_data, y_data, epochs=2000, batch_size=4)

#  .
print(model.weights)
print(" test data [5.,5.,5.]   : ", model.predict(test_data))
```
