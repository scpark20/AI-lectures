```python
import tensorflow as tf
## data 
x_data =[[1.],[2.],[3.],[4.]]
y_data =[[1.],[3.],[5.],[7.]]

test_data=[[5.]]

## tf.keras  perceptron  .
model = tf.keras.Sequential() ##    sequential  .      .
model.add(tf.keras.layers.Dense(1, input_dim=1)) #   add  . ,     1, perceptron 1.
model.summary() ##   

#  loss,   
optimizer=tf.keras.optimizers.SGD(lr=0.01)  ###   global min     .
loss=tf.keras.losses.mse  ##     . mse mean squre error ( - )^2  
metrics=tf.keras.metrics.mae ###     mse mean_absolute_error | - |  

#  
model.compile(loss=loss, optimizer=optimizer, metrics=[metrics])

#  
model.fit(x_data, y_data, epochs=500, batch_size=4)

#  .
print(model.weights)
print(" test data [5.]   : ", model.predict(test_data))
```
